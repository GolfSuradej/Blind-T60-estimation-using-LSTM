```python
import numpy as np
import scipy
import librosa

from scipy import signal

%matplotlib inline
import matplotlib.pyplot as plt
import thinkdsp
import thinkplot
```

```python
wavIn = thinkdsp.read_wave('TIMIT001.wav')
y_in = wavIn.ys
sr = wavIn.framerate
```

```python
y_re = librosa.resample(y_in, sr, 8000)
```

```python
x_in = np.linspace(0,1,len(y_in))
x_re = np.linspace(0,1,len(y_re))
plt.figure(figsize=(30,10))
plt.plot(x_in,y_in,'r-',x_re,y_re,'go')#,0.1,y[0],'ro')
plt.legend(['data', 'resampled'], loc='best')
plt.grid(True)
plt.axis((0.11, 0.12, -0.25, 0.25))
plt.show()
```
